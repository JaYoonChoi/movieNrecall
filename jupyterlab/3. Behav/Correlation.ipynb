```python
import pandas as pd
import numpy as np
import glob
import os 
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
file_dir = glob.glob(os.path.join('/mnt/c/Users/since/Dropbox/Lab/Projects/CJY/Baldassano clips/_DATA/rubrics/KClips'))
file = file_dir[0]
print(file)
survey_df = pd.read_excel(os.path.join(file,'MovieSurvey.xlsx'), sheet_name = 'survey')
```

```python
mean = survey_df['Sec_SecondRecall'].mean()
num = 0
for num in [0,1]:
    survey_df['Sec_SecondRecall'][num] = survey_df['Sec_SecondRecall'].mean()
    survey_df['Two_SecondRecall'][num] = survey_df['Two_SecondRecall'].mean()
    survey_df['King_SecondRecall'][num] = survey_df['King_SecondRecall'].mean()
    survey_df['Young_SecondRecall'][num] = survey_df['Young_SecondRecall'].mean()
    survey_df['Del_SecondRecall'][num] = survey_df['Del_SecondRecall'].mean()
    survey_df['Dan_SecondRecall'][num] = survey_df['Dan_SecondRecall'].mean()
    survey_df['Ava_SecondRecall'][num] = survey_df['Ava_SecondRecall'].mean()
    survey_df['Run_SecondRecall'][num] = survey_df['Run_SecondRecall'].mean()
```

```python
story = pd.DataFrame()
reinstatement  = survey_df.filter(regex='_reinstatement')
update = survey_df.filter(regex='_update')
FirstRecall = survey_df.filter(regex='_FirstRecall')
SecondRecall = survey_df.filter(regex='_SecondRecall')

story['reinstatement'] = reinstatement.mean(axis='columns')
story['diff'] = update.mean(axis='columns')
story['firstRecall'] = FirstRecall.mean(axis='columns')
story['secondRecall'] = SecondRecall.mean(axis='columns')

story.corr(method='pearson')
```
