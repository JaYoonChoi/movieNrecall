```python
#  
from scipy import stats
R = R1-R2
T = T1-T2
D = D1-D2
Y = Y1-Y2
sha, pvalue1 = stats.shapiro(R)
sha, pvalue2 = stats.shapiro(T)
sha, pvalue3 = stats.shapiro(Y) 
sha, pvalue4 = stats.shapiro(D)
print(pvalue1,pvalue2, pvalue3, pvalue4)
```

```python
# paired t test 
stat1,p_val1 = stats.ttest_rel(R1, R2)
print('Run', 'statistic:', stat1, '   p-value:', p_val1)
stat2,p_val2 = stats.ttest_rel(T1, T2)
print('Two', 'statistic:', stat2, '   p-value:', p_val2)
stat3,p_val3 = stats.ttest_rel(D1, D2)
print('Dance', 'statistic:', stat3, '   p-value:', p_val3)
stat4,p_val4 = stats.ttest_rel(Y1, Y2)
print('Young', 'statistic:', stat4, '   p-value:', p_val4)
```
