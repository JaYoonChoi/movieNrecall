```python
import numpy as np
import json
import csv
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
SG_dir = '/mnt/c/Users/since/Dropbox/RAs/projects/SangJun'
janny_dir = '/mnt/c/Users/since/Dropbox/jany/Movie_recall/_DATA/nlp/annotation_self_corr'
#recall_dir = os.path.join(SG_dir,'segmented','V11.json')
ref_dir = os.path.join(SG_dir,'annotation','reference2.json')
                    
with open(ref_dir, "r") as f:
    annot = json.load(f)
    #print(running_man)
annotations = {"RunningMan_integ" : annot['RunningMan_integ'], "RunningMan_audio" : annot['RunningMan_audio'], "RunningMan_visual" : annot['RunningMan_visual'], 
              "Dream_integ":annot['Dream_integ'], "Dream_audio":annot['Dream_audio'], "Dream_visual":annot['Dream_visual'],
              "Neighbors_integ":annot['Neighbors_integ'], "Neighbors_audio":annot['Neighbors_audio'], "Neighbors_visual":annot['Neighbors_visual'], 
              "BlindDate_integ":annot['BlindDate_integ'], "BlindDate_audio":annot['BlindDate_audio'], "BlindDate_visual":annot['BlindDate_visual'] }
programs = list(annotations.keys())
output_dir = os.path.join(janny_dir)

os.makedirs(output_dir, exist_ok=True)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS')
```

```python
programs
```

```python
from itertools import product
import numpy as np

all_result = {}
for program in programs:
    #result = {"annotations": annotations[program]}
    annotation_embeddings = model.encode(annotations[program])
    annotation_embeddings /= np.linalg.norm(annotation_embeddings, axis=-1, ord=2, keepdims=True)
    matrix = annotation_embeddings @ annotation_embeddings.T
    all_result[program] = matrix.tolist()

#with open(os.path.join(output_dir, "self_correlation.json"), "w") as f:
#    json.dump(all_result, f, indent=2, ensure_ascii=False)
```

```python
from itertools import product
import numpy as np

for story in ['RunningMan', 'Dream', 'Neighbors', 'BlindDate']: 
    audio = story+'_audio'
    visual = story+'_visual'
    condition = story+'_btw_cond'
    annotation_embeddings1 = model.encode(annotations[audio])   
    annotation_embeddings2 = model.encode(annotations[visual])   
    annotation_embeddings1 /= np.linalg.norm(annotation_embeddings1, axis=-1, ord=2, keepdims=True)
    annotation_embeddings2 /= np.linalg.norm(annotation_embeddings2, axis=-1, ord=2, keepdims=True)
    matrix = annotation_embeddings1 @ annotation_embeddings2.T
    all_result[condition]= matrix.tolist()

#with open(os.path.join(output_dir, "Dream_self_corr_btw_cond.json"), "w") as f:
#    json.dump(all_result_btw_cond, f, indent=2, ensure_ascii=False)
```

# Find event boundaries in the annotation  
- reference: https://github.com/brainiak/brainiak/blob/master/examples/eventseg/HiddenMarkovModels.ipynb

```python
from brainiak.eventseg.event import EventSegment
import matplotlib.patches as patches
```

```python
cond_boudary = {}
for program in programs:
    matrix = np.array(all_result[program])

    # find event segmentation 
    if program == 'Dream_integ'or program =='Dream_audio'or program =='Dream_visual'or  program =='Dream_btw_cond':
        ev = EventSegment(4) #segment events
    elif program == 'Neighbors_integ'or program =='Neighbors_audio'or program =='Neighbors_visual'or  program =='Neighbors_btw_cond':
        ev = EventSegment(6) #segment events
    else:
        ev = EventSegment(5) #segment events
    ev.fit(matrix)
    #print(program, ev)
    #get boundary 
    bounds = np.where(np.diff(np.argmax(ev.segments_[0], axis=1)))[0]
    cond_boudary[program] = bounds
    plt.figure(figsize=(6,6))

    plt.imshow(matrix)
    ax = plt.gca()
    bounds_aug = np.concatenate(([0],bounds, [53]))
    for i in range(len(bounds_aug)-1):
        rect = patches.Rectangle((bounds_aug[i],bounds_aug[i]),bounds_aug[i+1]-bounds_aug[i],bounds_aug[i+1]-bounds_aug[i],linewidth=1,edgecolor='w',facecolor='none')
        ax.add_patch(rect)
    plt.xlabel('Timepoints')
    plt.ylabel('Timepoints')
    plt.title(f"{program} Event Boundary", fontsize=12)
    #plt.savefig(f"{output_dir}/{program}_eb.png")
```

# Segmented_self_corr 

```python
print(np.array(all_result['RunningMan_visual']).shape)
print(np.array(all_result['RunningMan_audio']).shape)
print(np.array(all_result['RunningMan_integ']).shape)
print(np.array(all_result['RunningMan_btw_cond']).shape)
```

```python
eb_onset = {'RunningMan':[1,9,18,24,45], 'Neighbors':[1,6,18,21,31,43], 'Dream':[1,9,21,39], 'BlindDate':[1,14, 26, 40, 48]}
eb_offset = {'RunningMan':[9,18,24,45,55], 'Neighbors':[6,18,21, 31,43,55], 'Dream':[9,21,39,50], 'BlindDate':[14, 26, 40, 48,55]}

cond_seg_marixs, cond_seg_means, cond_means = {}, {}, {} 
for story in ['RunningMan', 'Dream', 'Neighbors', 'BlindDate']: #['RunningMan', 'Dream', 'Neighbors', 'BlindDate']:
    seg_matrixs, seg_means, mean, means = {}, {}, [], {}
    for cond in ['_visual', '_audio', '_btw_cond', '_integ']:
        program = story+cond
        matirx = pd.DataFrame(all_result[program]) 
        seg_matrix, seg_matrix_means = {}, {}
        for idx in range(len(eb_onset[story])):
            #print(story, cond, idx)
            matrix = matirx[eb_onset[story][idx]-1:eb_offset[story][idx]]
            matrix = matrix.loc[range(eb_onset[story][idx]-1,eb_offset[story][idx]-1), range(eb_onset[story][idx]-1,eb_offset[story][idx]-1)]
            seg_matrix[idx] = pd.DataFrame(matrix).sum(axis=1)/len(pd.DataFrame(matrix).columns)
            seg_matrix_means[idx] = (pd.DataFrame(seg_matrix[idx]).sum()/seg_matrix[idx].shape[0])[0]
        mean.append(sum(seg_matrix_means.values())/len(eb_onset[story]))
        #means[cond] = mean
        seg_matrixs[cond], seg_means[cond] =  seg_matrix, seg_matrix_means
    cond_seg_marixs[story], cond_seg_means[story], cond_means[story] = seg_matrixs, seg_means, mean
```

```python
# Effect of modality 
fig, axes = plt.subplots(1, 1, figsize = (5,5),sharey=True)   

mean_values_integ, mean_values_btw, mean_values_diff = [],[],[]
for col, story in enumerate(['RunningMan', 'Dream', 'Neighbors', 'BlindDate']):
    mean_values_integ.append(cond_means[story][3])
    mean_values_btw.append(cond_means[story][2])
    mean_values_diff.append(cond_means[story][0]-cond_means[story][1])
plt.bar(['RunningMan', 'Dream', 'Neighbors', 'BlindDate'], mean_values_btw, color='black', alpha=0.5, width=0.4)
plt.bar(['RunningMan', 'Dream', 'Neighbors', 'BlindDate'], mean_values_diff, color='blue', alpha=0.8, width=0.4)

#plt.bar(['RunningMan', 'Dream', 'Neighbors', 'BlindDate'],mean_values_btw, color='red', alpha=0.5, width=0.4)

plt.title('Diff', fontsize = 15)
plt.ylabel('R', fontsize = 15)
#axes[pos,col].axhline(0.0, 0, 1.0, color='black', linestyle='--', linewidth= 0.08)

fig.tight_layout()
plt.show()
```
