```python
run10_first = Run10[['V1','V2','V3','V4']].T
run10_second = Run10[['A1','A2','A3','A4','A5','A6','A7','A8']].T

# Summing all values along the vertical axis to get a (1, 50) Series
run10_first_series = run10_first.sum(axis=0)
run10_second_series = run10_second.sum(axis=0)
# If you need the result as a (1, 50) DataFrame instead of a Series:r
run10_first_df = run10_first_series.to_frame().transpose()
run10_second_df = run10_second_series.to_frame().transpose()

run10_change = run10_second_df- run10_first_df

# sns.set(rc={"figure.figsize": (6, 8)})
# sns.heatmap(run10_change.T)#, linewidth = .005)
# plt.title('S10_RunnignMan', fontsize=17)
run10_change = run10_change.T

#그래프생성
plt.rcParams["figure.figsize"] = (10,6)
#plt.plot(run10_change.index,run10_change[0],color='blue',linestyle='-',marker='o')
plt.bar(run10_change.index, run10_change[0])
#그래프 정보 설정
#plt.xlim(40, 80) #x축 범위
#plt.ylim(100, 200) #y축 범위
plt.xlabel('X') #x 라벨
plt.ylabel('Y') #y 라벨
plt.title("S10-RunningMan change") #그래프 이름

#그래프 출력
plt.savefig('S10_RunnignMan_change.png')
```
