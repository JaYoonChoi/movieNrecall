Arithmetic operations
The simplest operations we can perform is element-wise, what this means is that we want to perform some sort of mathematical operation on each voxel of the MR image. Since voxels are represented in a 3D array, this is equivalent to performing an operation on each element(i,j,k) of a 3D array. Let's try inverting the image, that is flip the colour scales such tat all blacks appear white and vice-versa. To do this, we'll use the method. 
img.math_img(formula, **imgs) where: 
- formula is a mathmetical expression such as 'a+1' 
- **imgs is a set of key-value pairs linking variable names to images. For example a=T1

In order to invert the image, we can simply flip the sign which will set the most positive elements (white) to the most negative elements(black), and the least positive elements (black) to the least negative elements (white). This effectively flips the colour-scale: 

```python
invert_img = img.math_img('-a', a=t1_img)
plot.plot_anat(invert_img)
```

```python
t1_test = img.math_img('a> 125', a=t1_img)
plot.plot_anat(t1_test)
```

```python
t1_test = img.math_img('a> 125', a=t1)
plot.plot_anat(t1_test)
```

- Applying a Mask 

Let's extend this idea of applying operations to each element of an image to multiple images. Instead of specifying just one image like the following: 
img.math_img('a+1', a= img_a) 
We can specifty multiple images by tacking on additional variables: 
img.math_img('a+b', a=img_a, b=img_b) 
The key requirement here is that when dealing with multiple images, that the size of the images must be same. The reason being is that we're dealing with element-wise operations. That means that some voxel (i.j.k) in img_a is being paired with some voxel in img_b when performing operations. So every voxel in img_a must have some pair with a voxel in img_b; size must be same. 
We can take advantage of this property when masking out data using multiplication. Masking works by multipling a raw imag (our t1), with some mask image (our bm). Whichever voxel (i,j,k) has a value of 0 in the mask multiplies with voxel (i,j,k) in the raw image resulting in a product of 0. <- the area which is outside of the brain has 0. 
Conversely, any voxel *(i,j,k) in the mask with a value of 1 multiplies with voxel (i,j,k) in the raw image resulting in the same value. Let's try this out in practice and see what the result is: 



```python
masked_t1 = img.math_img('a*b', a= t1, b=bm)
plot.plot_anat(masked_t1)
```

```python
inverted_mask = img.math_img('1-b', b=bm)
plot.plot_anat(inverted_mask)
```

```python
inverted_mask_t1 = img.math_img('a*b', a= t1, b=inverted_mask)
plot.plot_anat(inverted_mask_t1)
```

- Integrating Functional Data 

So far most of our work has been examining anatomical images -the reason being is taht it ptovides a nice visual way of exploring the effects of data manipulation and visualization is easy. In practice, you will most likely not analyze anatomical data using nilearn since there are other tools that are better suited for that kinds of analysis (-e.g.: freesurfer, connectome-workbench, mindboggle, etc..) 
In this note we wll finally start working with functional MR data-the modality of interest in this workshop. First we will cover some basics about how the data is organized (similar to T1s but more complex), and then how we can integrate our anatomical and fuctional data together using tools provided by nilearn. 

Functional data consists of full 3D brain volumes that are sampled at multiple time points. Therefore, you have a sequence of 3D brain volumes, stepping through sequences is stepping through time and therefore time is our 4th dimension. 

Each index along the 4th demensions (called TR for "Repetition Time", or Sample) is a full 3D scan of the brain. Pulling out volumes from 4-dimensional images is similar to that of 3-dimensional images except you're now dealing with 

img.slicer[x,y,z,time] 

Let's try a couple of examples to familiarize ourselves wirh dealing 4D images. But first, let's pull some functional data using PyBIDS.

- What fMRi actually represents

We've represented fMRi as an snapshot of MR signal over multipel timepoints. 
This is a useful way of understanding the organization of fMRI, however it isn't typically how we think about the data. Whe we analyze fMRI data. fMRI is typically thought of as time-series data. We can think of each voxel (x,y,z coordinate) as having a time-series of length T. The length T represents the number of volumes/timepoints in the data. Let's pick an example voxel and examine its time-series using func_mni_img.slicer

```python
 #looking a fluctuation of a single voxel accorss time-series
single_vox = func_mni_img.slicer[59:60,45:46,30:31,:].get_data()
```

```python
single_vox.shape
```

```python
single_vox = single_vox.flatten()
```

```python
single_vox.shape 
```

Here we've pulled out a voxel at a specific coordinate at every single time-point. This voxel has a single value for each timepoint and therefore is a time-series. We can visualize this time-series signal by using a standard plotting library. 

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(np.arange(0,single_vox.shape[0]), single_vox, 'k') # k means black
plt.xlabel('Timepoint')
plt.ylabel('Signal value')
```

# Resampling

- T1 image are typically composed of voxels that are 1x1x1 in dimension (fine resolution) 
- Fuctional images are typically composed of voxels that are 4x4x4 in dimension 

If we'd like to overlay out fuctional on top of our T1 (for visualization purpose or analyses), then we need to match the size of the voxel(resize). 

Note: Resampling is a method of interpolating in between data-points. When we stretch an image we need to figure out what goes in the spaces that are created via stretching - resampling dose just that. In fact, resizing any type of image is acutally just resampling to new dimensions. 

```python
#T1_mni = t1_files[0]
T1_mni = masked_t1
T1_mni_img = img.load_img(T1_mni)
```

```python
print(T1_mni_img.shape) 
print(func_mni_img.shape) # much smaller image so, need to stretch
```

A note on interpolation
1. continous
interpolate but maintain some edge features. Ideal for structual images where edges are well-defined. Uses 3rd order spline interpolation. 
2. Linear (default) 
Interpolate uses a combination of neighbouring voxels will blur. Uses trillnear interpolation. 
3. nearest
Mathes value of closest voxel(majority vote from neighbors). This is ideal for masks which are binary since it will preserve the 0's and 1's and will not produce in-between values (ex:0.342). Also ideal for numeric labels where values are 0,1,2,3 .. (parcellations). Uses nearest-neighbours interpolation with majority vote. 


```python
resamp_t1 = img.resample_to_img(source_img=T1_mni_img, target_img = func_mni_img, 
                                 interpolation='continuous')
print(resamp_t1.shape)
print(func_mni_img.shape)
plot.plot_anat(resamp_t1)
```

# Brain Atlases

A brain atlas is a labeling of a volume. Each voxel constains a label, instead of a gray value. These labels are numeric. Each number corresponds to a region of interest(ROI). 

Why ROIs? 

1. Explore the data (see signal in areas of interest plotted for each condition or plotted against other variables of interest) 
2. Limit number of statistical tests, controlling for Type 1 Error 
3. Limit statistical tests to a regions that is functionally defined. 

compareing Atlases: See Arisan et al., 2018
