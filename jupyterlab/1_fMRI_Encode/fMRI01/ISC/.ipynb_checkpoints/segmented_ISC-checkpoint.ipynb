```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_rel
from sklearn.metrics import pairwise_distances
import pandas as pd

def bootstrap_subject_matrix(similarity_matrix, random_state=None):
    ''' Shuffle subjects within a similarity matrix based on recommendation by Chen et al., 2016 '''
    random_state = np.random.RandomState(random_state)
    n_sub = similarity_matrix.shape[0]
    bootstrap_subject = sorted(random_state.choice(np.arange(n_sub), size=n_sub, replace=True))
    return similarity_matrix[bootstrap_subject, :][:, bootstrap_subject]

def create_null_distribution(isc_values, n_bootstraps=1000):
    ''' Create a null distribution by bootstrapping ISC values '''
    null_distribution = []
    for _ in range(n_bootstraps):
        # Bootstrap the ISC values
        bootstrapped_isc = bootstrap_subject_matrix(isc_values)
        # Calculate the mean ISC for the bootstrapped sample
        mean_isc = np.mean(bootstrapped_isc)
        null_distribution.append(mean_isc)
    return np.array(null_distribution)

# Assuming isc_session1 and isc_session2 are your ISC matrices from two encoding sessions
# For example purposes, we create some random data
isc_session1 = 1 - pairwise_distances(pd.DataFrame(np.random.rand(10, 5)).T, metric='correlation')
isc_session2 = 1 - pairwise_distances(pd.DataFrame(np.random.rand(10, 5)).T, metric='correlation')

# Create null distributions for each session
null_distribution1 = create_null_distribution(isc_session1, n_bootstraps=5000)
null_distribution2 = create_null_distribution(isc_session2, n_bootstraps=5000)

# Calculate the observed difference in mean ISC
observed_diff = np.mean(isc_session1) - np.mean(isc_session2)

# Combine the null distributions and calculate p-value
combined_null = null_distribution1 - null_distribution2
p_value = np.mean(np.abs(combined_null) >= np.abs(observed_diff))

# Output results
print(f"Observed difference in ISC: {observed_diff}")
print(f"P-value: {p_value}")

# Plotting
plt.figure(figsize=(12, 6))
sns.histplot(combined_null, kde=True, color='blue', label='Null Distribution of ISC Difference')
plt.axvline(observed_diff, color='red', label='Observed Difference')
plt.xlabel('Difference in ISC')
plt.ylabel('Frequency')
plt.title('Null Distribution vs. Observed Difference in ISC')
plt.legend()
plt.show()
```

```python
# Calculate inter-SC Sliding window 

import pandas as pd
from tqdm import tqdm

#roi =28 
#       
window_size = 20 #TR - 60secs 
stride = 20  #    (TR )

InterSC_r, InterSC_p = {}, {}
for roi in [5,19,28,32]:
    story_ISCr, story_ISCp = {}, {}
    for story in tqdm(['run', 'blind', 'dream', 'nei']):
        cond_ISCr, cond_ISCp = {}, {}
        for cond in ['visual', 'audio']:
            data = pd.DataFrame(story_run_subjs_ROIs[roi][story][cond])
            window_isc_r, window_isc_p = [],[]
            for i in range(0, data.shape[0] - window_size + 1, stride):
                window = data.iloc[i:i + window_size]
                stats = isc(window, n_bootstraps=10000, metric='median', method='bootstrap')
                window_isc_r.append(stats['isc'])
                window_isc_p.append(stats['p'])
            cond_ISCr[cond] = pd.Series(window_isc_r, index=data.index[window_size - 1::stride])
            cond_ISCp[cond] = pd.Series(window_isc_p, index=data.index[window_size - 1::stride])
        story_ISCr[story], story_ISCp[story] = cond_ISCr, cond_ISCp
    InterSC_r[roi], InterSC_p[roi]= story_ISCr, story_ISCp
```

## Sliding Intra-SC

```python
# Calculate intra-SC Sliding window 

import pandas as pd
from tqdm import tqdm

window_size = 15 #TR - 120secs  #       
stride = 15  #    (TR )

Intra_SCr,Intra_SCp = {}, {}
for roi in tqdm([5,19, 28, 32]):
    sub_ISCr, sub_ISCp = {},{}
    for sub in subjects:
        sliding_r, sliding_p = {},{}
        for story in ['run', 'blind', 'dream', 'nei']:
            data = pd.DataFrame(ROIs[roi][sub][story])
            window_isc_r, window_isc_p = [], []
            for i in range(0, data.shape[0] - window_size + 1, stride):
                window = data.iloc[i:i + window_size]
                stats = isc(window, n_bootstraps=10000, metric='median', method='bootstrap', exclude_self_corr=True) #Same Subject multiple repeat option : exclude_self_corr=True
                window_isc_r.append(stats['isc'])
                window_isc_p.append(stats['p'])
            sliding_r[story] = pd.Series(window_isc_r, index=data.index[window_size - 1::stride])
            sliding_p[story] = pd.Series(window_isc_p, index=data.index[window_size - 1::stride])
        sub_ISCr[sub],sub_ISCp[sub]  = sliding_r, sliding_p
    Intra_SCr[roi], Intra_SCp[roi] = sub_ISCr, sub_ISCp
```

```python
# get mean patterns for each roi & stories 

means = {} 
for roi in tqdm([5,19, 28, 32]):
    story_means = {}
    for story in ['run', 'blind', 'dream', 'nei']:
        mean_df = {} 
        for sub in subjects:
            mean_df[sub] = Intra_SCr[roi][sub][story]
        story_means[story] = pd.DataFrame(mean_df).sum(axis=1)/len(pd.DataFrame(mean_df).columns)
    means[roi] =  story_means
```
