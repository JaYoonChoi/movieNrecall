```python
def load_masker():
    
    note_path = os.getcwd()
    mask_path = os.path.join(note_path,'roi_mask')

    # read roi_mask files in the folder
    mask_files = os.path.join(mask_path,'comb','roi_[cp]*_*.nii.gz')
    roi_mask_files = glob(mask_files)
    sorted(roi_mask_files)

    # create masker
    masker_dict = {}

    for idx_roi, this_roi in enumerate(sorted(roi_mask_files)):
        pattern = "comb/roi_(.*?)_mask.nii.gz"
        roi_name = re.search(pattern, this_roi).group(1)

        masker_dict[roi_name] = NiftiMasker(this_roi)
    return masker_dict
```

## Extract Time Course for the trial-type specific modulations

```python
def load_events(layout, subject, this_task, n):  
    import pandas as pd

    event_filenames = layout.get(scope = 'raw',
                                  subject = subject,
                                  task = this_task,
                                  suffix = 'events')
    events = pd.DataFrame()

    for event_filename in event_filenames:

        event = pd.read_csv(event_filename, sep = '\t', 
                            usecols = ['onset','duration','cued_value','image_condition','gamble_choice','response_time']) 

        run_label = event_filename.entities.get('run')
        event['run'] = run_label

        event.index.names = ['run_trial']
        event.reset_index(drop = False, inplace = True)

        event['trial'] = event['run_trial'] + len(event['run_trial'])*(run_label-1)
        event['trial_type'] = event['gamble_choice'] + event['image_condition']

        event['prev_condition'] = [np.nan, *event.image_condition[:-1]]
        event['prev_choice'] = [np.nan, *event.gamble_choice[:-1]]
        event['next_choice'] = [*event['gamble_choice'][1:], np.nan]

        event['current_stay'] = event['gamble_choice']==event['prev_choice']
        event['next_stay'] = event['gamble_choice']==event['next_choice']

        labels = {True:'stay', False:'switch'}
        event['current_stay'].replace(labels, inplace=True)
        event['current_stay'].iloc[0] = np.nan
        event['next_stay'].replace(labels, inplace=True)
        event['next_stay'].iloc[-1] = np.nan
        labels = {'acceptgain':'good', 'rejectloss':'good',
                  'acceptloss':'bad',  'rejectgain':'bad'}

        event['pe'] = event['trial_type']
        event['pe'].replace(labels, inplace=True)
        event['prev_pe'] = [np.nan, *event['pe'][:-1]]
        event['condition'] = event['current_stay']+event['pe']        

        # calculate the probability of the previous trials 
        # if N is set, calculate only the previous N trials
        
        nprob_this = np.full((42,1),0.0)
        nprob_prev = np.full((42,1),0.0)

        for this_index, this_event in enumerate(event['image_condition']):
            this_sequence = []
            for i1 in range(n):
                if this_index - n + i1 >= 0:
                    this_sequence.append(this_index - n + i1)

            this_cards = event['image_condition'][this_sequence]
            this_cards[this_cards == 'gain'] = 1
            this_cards[this_cards == 'loss'] = 0


            nprob_this[this_index] = expected_prob(this_cards, n, 10)

        event['est_prob'] = nprob_this

        events = pd.concat([events, event])

    orig_events = copy.deepcopy(events)
    
    this_rows = events.loc[events['gamble_choice'] == 'na'].index.values
    events.drop(events.index[this_rows], inplace = True)
    events.set_index(['trial'],inplace = True)
    events['subject'] = subject
    
    
    return events, orig_events
```

```python
def expected_prob(this_cards, n, m):

    post = np.full((m+1),1.0)
    x = np.arange(0,m+1)/m
    y = 1-x;
    z = np.multiply(x,y)
    
    if len(this_cards) < n:
        for i2 in range(n-len(this_cards)):
            post = np.multiply(post,z)
       
        for i2 in this_cards:
            if i2 == 1: # gain
                post = np.multiply(post,x)
            else:
                post = np.multiply(post,y)
    else:
        for i2 in this_cards:
            if i2 == 1: # gain
                post = np.multiply(post,x)
            else:
                post = np.multiply(post,y)
        
    return np.sum(np.multiply(x,post/np.sum(post))) # return expected probability
```

```python
def load_nii(file_name):
    
    nii = []
    import nibabel as nib
    for this_file in file_name:
        nii.append(nib.load(this_file))
        
    return nii
```

```python
def load_filelist(subject,task):
    
    user_path = os.path.expanduser('~')
    home_path = os.path.join(user_path,'Desktop')

    if task == 'gamble':
        prep_path = os.path.join(home_path,'gamble_comb','Gamble',
                                 'data','derivatives')
    else:
        prep_path = os.path.join(home_path,'gamble_comb','GambleControl',
                                 'data','derivatives')
        
    
    from glob import glob

    n_task = 4

    filenames = os.path.join(prep_path,
                              'sub-{}'.format(subject),
                              'func',
                              'sub-{subject}_task-{task}_run-[0-9]_space-MNI152NLin6Asym_desc-smoothAROMAnonaggr_bold.nii.gz'.format(
                              subject = subject,
                              task = task))


    func_filenames = glob(filenames)
    func_filenames = sorted(func_filenames)
    
    return func_filenames
```

```python
def load_events_wo_layout(subject, task, n):
    
    import pandas as pd
    import glob

    user_path = os.path.expanduser('~')
    home_path = os.path.join(user_path,'Desktop')

    if task == 'gamble':
        data_path = os.path.join(home_path,'gamble_comb','Gamble',
                                 'data')
        
        templ_filenames = os.path.join(data_path,
                                       'sub-{}'.format(subject),
                                       'func',
                                       'sub-{subject}_task-{task}_run-0[0-9]_events.tsv'.format(
                                       subject = subject,
                                       task = task))
        
    else:
        data_path = os.path.join(home_path,'gamble_comb','GambleControl',
                                 'data')
        
        templ_filenames = os.path.join(data_path,
                                       'sub-{}'.format(subject),
                                       'func',
                                       'sub-{subject}_task-{task}_run-[0-9]_events.tsv'.format(
                                       subject = subject,
                                       task = task))

    event_filenames = sorted(glob.glob(templ_filenames))
    events = pd.DataFrame()

    for event_filename in event_filenames:
        event = pd.read_csv(event_filename, sep = '\t', 
                            usecols = ['onset','duration','cued_value','image_condition','gamble_choice','response_time']) 
        
        if task == 'gamble':
            run_label = int(re.search('task-gamble_run-(.*)_events',event_filename).group(1))
        else:
            run_label = int(re.search('task-gamblecontrol_run-(.*)_events',event_filename).group(1))
        event['run'] = run_label

        event.index.names = ['run_trial']
        event.reset_index(drop = False, inplace = True)

        event['trial'] = event['run_trial'] + len(event['run_trial'])*(run_label-1)
        event['trial_type'] = event['gamble_choice'] + event['image_condition']

        event['prev_condition'] = [np.nan, *event.image_condition[:-1]]
        event['prev_choice'] = [np.nan, *event.gamble_choice[:-1]]
        event['next_choice'] = [*event['gamble_choice'][1:], np.nan]

        event['current_stay'] = event['gamble_choice']==event['prev_choice']
        event['next_stay'] = event['gamble_choice']==event['next_choice']

        labels = {True:'stay', False:'switch'}
        event['current_stay'].replace(labels, inplace=True)
        event['current_stay'].iloc[0] = np.nan
        event['next_stay'].replace(labels, inplace=True)
        event['next_stay'].iloc[-1] = np.nan
        labels = {'acceptgain':'good', 'rejectloss':'good',
                  'acceptloss':'bad',  'rejectgain':'bad'}

        event['pe'] = event['trial_type']
        event['pe'].replace(labels, inplace=True)
        event['prev_pe'] = [np.nan, *event['pe'][:-1]]
        event['condition'] = event['current_stay']+event['pe']        

        # calculate the probability of the previous trials 
        # if N is set, calculate only the previous N trials
        
        nprob_this = np.full((42,1),0.0)
        nprob_prev = np.full((42,1),0.0)

        for this_index, this_event in enumerate(event['image_condition']):
            this_sequence = []
            for i1 in range(n):
                if this_index - n + i1 >= 0:
                    this_sequence.append(this_index - n + i1)

            this_cards = event['image_condition'][this_sequence]
            this_cards[this_cards == 'gain'] = 1
            this_cards[this_cards == 'loss'] = 0


            nprob_this[this_index] = expected_prob(this_cards, n, 10)

        event['est_prob'] = nprob_this

        events = pd.concat([events, event])

    orig_events = copy.deepcopy(events)
    
    this_rows = events.loc[events['gamble_choice'] == 'na'].index.values
    events.drop(events.index[this_rows], inplace = True)
    events.set_index(['trial'],inplace = True)
    events['subject'] = subject
    
    
    return events, orig_events
```

```python
def distill_event(subject, task):
    
    n = 10
    events, orig_event = load_events_wo_layout(subject, task, n)  
    
    target = ['image_condition','gamble_choice','prev_condition','prev_choice']
    
    drops_ = np.full(events.shape[0],False)
    for this_target in target:
        drops = events[this_target].isna()
        drops_ = drops_ | drops 

    for problem in ['nagain','naloss']:
        drops = events['pe'] == problem
        drops_ = drops_ | drops 

    temp_labels = events[target][np.logical_not(drops_)]
    labels = temp_labels.values
    new_events = events[:][np.logical_not(drops_)]
    
    return new_events, labels
```

```python
def extract_timeseries(subject_task):
    
    ts_len = 15
    offset = 5 # if set 0, segment from the onset of the image onset
    
    subject = subject_task[0]
    task = subject_task[1]
    
    masker_dict = load_masker()
    file_list = load_filelist(subject, task)
    events, labels = distill_event(subject, task)
    nii = load_nii(file_list)
    
    all_series = []
    for this_masker in masker_dict: # for multiple masker, this is not a good loop
        print('sub-' + subject + '_task-' + task + '_masker-' + this_masker + '\n')
        tsfmri_ = []

        for this_run in range(4):
            print('sub-' + subject + '_run-' + str(this_run))
            this_events = events.loc[events['run'] == this_run+1]
            this_func = nib.load(file_list[this_run])

#             fmri_masked = masker_dict[this_masker].fit_transform(this_func)
#             fmri_masked_pad = np.pad(fmri_masked,((0,20),(0,0)),'mean')
#             fmri_masked_pad = np.mean(fmri_masked_pad,axis=1)

#             tsfmri = np.full((this_events.shape[0]),0.0)
            
            pad_size = 10;
            fmri_masked = np.mean(masker_dict[this_masker].fit_transform(this_func),axis=1)
            fmri_masked_pad = np.pad(fmri_masked,(pad_size,pad_size),'mean')

            tsfmri = np.full((this_events.shape[0],ts_len),0.0)
            
            for idx, this_onset in enumerate(this_events['onset']):
                vol = int(this_onset/2)-offset + pad_size
                this_segment = fmri_masked_pad[[vol + i1 for i1 in range(ts_len)]]
                tsfmri[idx,:] = this_segment

            tsfmri_.append(tsfmri)

        fmri_series = np.vstack(tsfmri_)


        all_series.append([fmri_series, labels])
        
    return all_series
```

```python
def run_trial_average(results, bcor, which_masker):
    
    masker_dict = load_masker()

    for idx_masker, this_masker in enumerate(masker_dict.keys()):
        if this_masker == which_masker:
            which_idx = idx_masker

    masked_mean = []
    for subj_result in results:
        this_masked = subj_result[which_idx][0]

        dfl = pd.DataFrame(subj_result[which_idx][1],columns=['image','choice','prev_image','prev_choice'])

        ts_len = 15 # this_masked.shape[1]
        offset = 5

        if bcor == 'onset':
            # baseline correction at the stimulus onset
            normalized_ni = (this_masked-this_masked[:,offset][:,None])/this_masked[:,offset][:,None]
        else:
            # baseline correction at the decision window
            normalized_ni = (this_masked-np.mean(this_masked[:,[offset-1,offset-0]],axis=1)[:,None])/np.mean(this_masked[:,[offset-1,offset-0]],axis=1)[:,None]


        bold_mean = np.full((4,ts_len),0.0)


        for idx_pair, this_pair in enumerate(product(['accept','reject'],['gain','loss'])):

            trial_index = dfl.loc[(dfl['choice'] == this_pair[0]) & 
                                  (dfl['image']  == this_pair[1])].index

            if len(trial_index) > 0:
                bold_mean[idx_pair,:] = np.nanmean(normalized_ni[trial_index,:],axis=0)
            else:
                bold_mean[idx_pair,:] = 0*np.arange(0,ts_len-1)
        masked_mean.append(bold_mean)
    
    return masked_mean
```

```python
# plot the results
masker_dict = load_masker()
bcor = 'mean'
fig, ax = plt.subplots(len(masker_dict),4)
fig.set_size_inches(20, 15)


cline = ['steelblue','lightcoral','darkblue','maroon']
sline = ['solid','solid','dashed','dashed']

for idx_masker, this_masker in enumerate(masker_dict.keys()):
    print(this_masker)
    for idx_plot, this_cond in enumerate(['gamble','gamble_control']):
           
        if this_cond == 'gamble':
            output = run_trial_average(results, bcor, this_masker)
        else:
            output = run_trial_average(results_control, bcor, this_masker)
                                         
        
        mean_bold = np.full((len(output),4,15),0.0)
        for idx, this in enumerate(output):
            mean_bold[idx,:,:] = this

        for idx_pair, this_pair in enumerate(product(['accept','reject'],['gain','loss'])):
            if idx_pair < 2:
                bold_interest = mean_bold[:,idx_pair,:]
                idx_plus = idx_pair
            else:
                bold_interest = mean_bold[:,idx_pair,:]
                idx_plus = idx_pair-2
            
            if idx_masker == 0:
                print(this_cond)
                print(this_pair)
                
            ax[idx_masker,2*idx_plot+idx_plus].errorbar(np.array([i for i in range(15)]),
                                    100*np.nanmean(bold_interest,axis=0),
                                    100*np.nanstd(bold_interest,axis=0)/np.sqrt(bold_interest.shape[0]),
                                    color=cline[idx_pair],
                                    linestyle =sline[idx_pair],
                                   )
            ax[idx_masker,2*idx_plot+idx_plus].set_ylim((-0.15, 0.15))
        
        ax[0,0].legend(['acc-gain','acc-loss'])
        ax[0,1].legend(['rej-gain','rej-loss'])
        ax[0,2].legend(['acc-gain','acc-loss'])
        ax[0,3].legend(['rej-gain','rej-loss'])
```
